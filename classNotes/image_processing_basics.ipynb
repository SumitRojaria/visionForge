```python
#Importing Cv2
import cv2
from google.colab.patches import cv2_imshow
```

```python
# Mounting google drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
image_path = '/content/drive/MyDrive/sample_image.jpeg'
#Reading the image
img = cv2.imread(image_path)
#Displaying the image
cv2_imshow(img)
```

```python
#Converting to gray scale
img1 = cv2.imread(image_path,0)
cv2_imshow(img1)
```

```python
#saving the image
cv2.imwrite('gray_scale.png', img1)
```

```python
#Converting Image to Array
from numpy import asarray
arr = asarray(img)
print(arr)
```

```python
#Rotate Image
import imutils
r1 = imutils.rotate(img,45)
cv2_imshow(r1)
```

```python
import imutils
r1 = imutils.rotate(img,90)
cv2_imshow(r1)
```

```python
import imutils
r1 = imutils.rotate(img,180)
cv2_imshow(r1)
```

```python
#converting image into RGB components
R,G,B=cv2.split(img)
cv2_imshow(R)
cv2_imshow(G)
cv2_imshow(B)
```

```python
#converting image into CMY components
C=1-R
M=1-G
Y=1-B
cv2_imshow(C)
cv2_imshow(M)
cv2_imshow(Y)
```
